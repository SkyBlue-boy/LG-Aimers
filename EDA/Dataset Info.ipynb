제품 : 제품 코드
대분류 : 제품의 대분류 코드
중분류 : 제품의 중분류 코드

브랜드 : 제품의 브랜드 코드

단, 제품이 동일하여도 판매되고 있는 고유 ID 별로 기재한 분류 정보가 상이할 수 있음
